```python
import pandas as pd
import numpy as np
```

```python
df_bert = pd.read_pickle("support_filtered.pkl")
df_bert
```

```python
df_jaccard = pd.read_pickle("jaccard.pkl")
df_jaccard = df_jaccard.rename(columns={'support': 'jaccard'})
df_jaccard
```

```python
df_jaccard['bert'] = df_bert.loc[:, ('support')]
df_jaccard.to_pickle("final.pkl")
df_jaccard
```

```python
df = pd.read_pickle("final.pkl")
bert_dict = df.loc[:, ('jaccard')]
jaccard_dict = df.loc[:, ('bert')]


overlap_ratio = 0

overlaps = np.array([])

cnt = 0

for idx in range (len(bert_dict)):
    bert_keys = list(bert_dict[idx].keys())
    jaccard_keys = list(jaccard_dict[idx].keys())
    for elem in jaccard_keys:
        if cnt <= 20:
            if elem in bert_keys:
                overlap_ratio = overlap_ratio + 1
        cnt = cnt + 1
    overlaps = np.append(overlaps, overlap_ratio/19, axis=None)
    overlap_ratio = 0
    cnt = 0
df['number of jaccards seen in bert/total number of berts'] = overlaps
print(overlaps)
print(sum(overlaps)/len(overlaps))
df.to_pickle("final_scoresim.pkl")
```
